```python
import json

class Employee:
    def __init__(self, name, dob, height, city, state):
        self.name = name
        self.dob = dob
        self.height = height
        self.city = city
        self.state = state

    def __str__(self):
        return f"Name: {self.name}, DOB: {self.dob}, Height: {self.height}, City: {self.city}, State: {self.state}"

employee_data = [
    {
        "Name": "John Doe",
        "DOB": "1990-01-15",
        "Height": 175,
        "City": "New York",
        "State": "New York"
    },
    {
        "Name": "Alice Smith",
        "DOB": "1985-05-20",
        "Height": 160,
        "City": "Los Angeles",
        "State": "California"
    },
    {
        "Name": "Bob Johnson",
        "DOB": "1992-09-03",
        "Height": 182,
        "City": "Chicago",
        "State": "Illinois"
    },
    {
        "Name": "Eva Brown",
        "DOB": "1988-03-10",
        "Height": 168,
        "City": "Houston",
        "State": "Texas"
    },
    {
        "Name": "David Lee",
        "DOB": "1995-07-25",
        "Height": 178,
        "City": "San Francisco",
        "State": "California"
    }
]

with open("employee.json", "w") as file:
    json.dump(employee_data, file)

employee_objects = []
with open("employee.json", "r") as file:
    data = json.load(file)
    for item in data:
        employee = Employee(item["Name"], item["DOB"], item["Height"], item["City"], item["State"])
        employee_objects.append(employee)

for employee in employee_objects:
    print(employee)
```

```python
class Dog:
    def __init__(self, name, age, coat_color):
        self.name = name
        self.age = age
        self.coat_color = coat_color

    def description(self):
        print(f"{self.name} is {self.age} years old.")

    def get_info(self):
        print(f"{self.name}'s coat color is {self.coat_color}.")


class JackRussellTerrier(Dog):
    def __init__(self, name, age, coat_color, special_ability):
        super().__init__(name, age, coat_color)
        self.special_ability = special_ability

    def description(self):
        print(f"{self.name} is a Jack Russell Terrier and is {self.age} years old.")

    def bark(self):
        print(f"{self.name} is barking loudly!")

    def perform_trick(self):
        print(f"{self.name} is performing a trick: {self.special_ability}")


class Bulldog(Dog):
    def __init__(self, name, age, coat_color, strength):
        super().__init__(name, age, coat_color)
        self.strength = strength

    def description(self):
        print(f"{self.name} is a Bulldog and is {self.age} years old.")

    def snore(self):
        print(f"{self.name} is snoring loudly!")

    def show_strength(self):
        print(f"{self.name} has a strength rating of {self.strength}.")


# Create Dog objects
dog1 = Dog("Rex", 5, "Brown")
dog2 = JackRussellTerrier("Buddy", 3, "White and Brown", "Jumping through hoops")
dog3 = Bulldog("Max", 4, "Brindle", "High")

# Calling methods on Dog objects
dog1.description()
dog1.get_info()

# Calling methods on JackRussellTerrier objects
dog2.description()
dog2.get_info()
dog2.bark()
dog2.perform_trick()

# Calling methods on Bulldog objects
dog3.description()
dog3.get_info()
dog3.snore()
dog3.show_strength()
```
